# Εγγραφή `(subscribe)` στον `mtqtt broker` για την λήψη μηνυμάτων που έχουν σταλεί

```python
#@title Μέθοδος 1η: Ατέρμονη επανάληψη για τη δημιουργία μόνιμης σύνδεσης με mqtt broker { vertical-output: true }
broker_address="broker.hivemq.com" #@param {type:"string"}
topic = "weather_data" #@param {type:"string"}

# -- Εισαγωγή βιβλιοθηκών --------------------------------------------------
import paho.mqtt.client as mqtt
import time

# -- Συνάρτηση ΧΧΧΧ --------------------------------------------------------
def on_message(client, userdata, message):
    print("message received " ,str(message.payload.decode("utf-8")))
    print("message topic=",message.topic)

# -- Σύνδεση σε broker και topic --------------------------------------------
print("creating new instance")
client = mqtt.Client("") #create new instance
client.on_message=on_message #attach function to callback
print("connecting to broker")
client.connect(broker_address) #connect to broker
client.loop_start() #start the loop
print("Subscribing to topic",topic)
client.subscribe(topic)
client.loop_forever() # useful when the program must run indefinitely.
                      # loop_forever() function also handles automatic reconnects.
```

```python
#@title Μέθοδος 2η: Δημιουργία ενός εξομοιωτή terminal με χρήση `IPython` και `HTML`
#@markdown Στο terminal που θα προκύψει θα χρησιμοποιήσουμε τα προγράμματα `mosquitto` και `mosquitto_clients` για να μπορέσουμε να κάνουμ `subscribe` στον broker
# ------------------------------------------------------------------------------
from IPython.display import JSON
from google.colab import output
from subprocess import getoutput
import os

def shell(command):
  if command.startswith('cd'):
    path = command.strip().split(maxsplit=1)[1]
    os.chdir(path)
    return JSON([''])
  return JSON([getoutput(command)])
output.register_callback('shell', shell)
```
